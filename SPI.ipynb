```python
file=r'/Users/wenqi/OneDrive - University of Nebraska-Lincoln/Sharedbyme_Wenqi/2st_paper/Data/Drought Index/SPI_EWT.xlsx'
ewt=pd.read_excel(file,sheet_name="Sheet2")
ewt['Date'] = pd.to_datetime(ewt['Date'], format='%Y%m%d')
ewt["Date_2"]=ewt["Date"].dt.year.astype(str)+ewt["Date"].dt.week.astype(str)
ewt_0=ewt.groupby("Date_2").mean().reset_index()
ewt_0["Date_3"]=ewt_0["Date_2"].astype(int)
ewt_0["Date_3"]=ewt_0["Date_3"].astype(str)
ewt_0
```

```python
def stat(x,y):
    import numpy as np
    from sklearn.linear_model import LinearRegression
    from sklearn.metrics import mean_squared_error
    X=[[i]for i in x]
    model = LinearRegression(fit_intercept=True)
    model.fit(X, y)
    xfit = x
    yfit = model.predict(X)
    r_squared=model.score(X,y)
    mean_squared_error(yfit, xfit)
    rmse = np.sqrt(mean_squared_error(yfit, y))
    model.score(X,y)
    nrmse=rmse/(max(y)-min(y))
    return [r_squared,rmse,yfit,X,nrmse]
x=ewt_0["Z_Score"]
y=data_2_0.loc[ewt_0["Date_3"],"SPI"]
stat(x,y)[0]
```
